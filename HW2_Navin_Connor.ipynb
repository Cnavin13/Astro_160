What is a key property of the T evolution while the star is on the main sequence (MS), compared to the phases before and after?

```python
star_age0 = eep0.eeps['star_age']
cent_temp0 = eep0.eeps['log_center_T']
cent_dense0 = eep0.eeps['log_center_Rho']
Teff0 = eep0.eeps['log_Teff']

star_age1 = eep1.eeps['star_age']
cent_temp1 = eep1.eeps['log_center_T']
cent_dense1 = eep1.eeps['log_center_Rho']
Teff1 = eep1.eeps['log_Teff']

star_age2 = eep2.eeps['star_age']
cent_temp2 = eep2.eeps['log_center_T']
cent_dense2 = eep2.eeps['log_center_Rho']
Teff2 = eep2.eeps['log_Teff']

star_age3 = eep3.eeps['star_age']
cent_temp3 = eep3.eeps['log_center_T']
cent_dense3 = eep3.eeps['log_center_Rho']
Teff3 = eep3.eeps['log_Teff']

log = np.log10
```

```python
plt.plot(log(star_age0), log(cent_temp0), linestyle = 'dashed')
plt.plot(log(star_age1), log(cent_temp1), linestyle = 'dotted')
plt.plot(log(star_age2), log(cent_temp2), linestyle = 'dashdot')
plt.plot(log(star_age3), log(cent_temp3), linestyle = 'solid')

plt.title('HR plot Age v Temp', fontsize=20)
plt.ylabel('Star Age(yr)', fontsize=15)
plt.xlabel('Central Temperature(K)', fontsize=15)
plt.legend(fontsize=10)
plt.show()
```
